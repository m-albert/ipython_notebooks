## Manipulating arrays

```python
# let's have another array:
llexpStemcB = []
lgnStemcB = []
with open(fnStemcExpB, 'r') as f:
    lHdrB = f.readline().strip().split('\t')[2:]
    for l in f:
        l = l.strip().split('\t')
        lgnStemcB.append(l[0])
        llexpStemcB.append([float(x) for x in l[2:]])

aexpStemcB = np.array(llexpStemcB)
```

```python
aexpStemcAB = np.vstack((aexpStemc, aexpStemcB))
```

```python
# freeing up memory:
del aexpStemcAB
del aexpStemcB
```

```python
# adding one column
aexpStemcPlus1 = np.zeros((aexpStemc.shape[0], aexpStemc.shape[1] + 1), dtype = np.float64)
aexpStemcPlus1[:,:-1] = aexpStemc
aexpStemc = aexpStemcPlus1
del aexpStemcPlus1
```

```python
aexpStemc[:,-1] = aexpStemc.mean(axis = 1)
aexpStemc[:,-1] = np.nanmean(aexpStemc, axis = 1)

# ordering the rownames first
agnStemc = agnStemc[aexpStemc[:,-1].argsort()]
aexpStemc = aexpStemc[aexpStemc[:,-1].argsort(),:]
```

```python
# closest value to the mean?
m = np.nanmean(aexpStemc[:,:-1])
# index of the upper closest number:
iuMean = aexpStemc[:,-1].searchsorted(m)
```

```python
print 'mean = %.04f, closest upper = %.04f, closest lower = %.04f' % (iuMean, aexpStemc[iuMean,-1], aexpStemc[iuMean - 1,-1])
```

```python
# re-creating the array
aexpStemc = np.array(llexpStemc)
aHdr = np.array(lHdr)
agnStemc = np.array(lgnStemc)
```

```python
# getting a smaller array: the 100 highest expressed genes:
agnStemc = agnStemc[aexpStemc.mean(axis = 1).argsort()]
aexpStemc = aexpStemc[aexpStemc.mean(axis = 1).argsort(),:]
```

```python
aexpTop100 = aexpStemc[:100,:]
```

## Igraph

```python
acorTop100 = np.array([stats.stats.pearsonr(aexpTop100[i,:], aexpTop100[j,:])[0] \
     for i in xrange(aexpTop100.shape[0]) \
     for j in xrange(aexpTop100.shape[0])])
```

```python
acorTop100.shape = (aexpTop100.shape[0], aexpTop100.shape[0])
```

```python
acorTop100.shape
```

```python
acorTop100
```

```python
acorTop100 = np.triu(acorTop100)
```

```python
acorTop100[np.arange(acorTop100.shape[0]), np.arange(acorTop100.shape[1])] = 0.0
```

```python
acorTop100
```

```python
p97 = np.percentile(np.abs(acorTop100), 97.0)
```

```python
tlGenesCor = [(agnStemc[i], agnStemc[j], acorTop100[i,j]) \
    for i in xrange(100) \
    for j in xrange(100) \
    if np.abs(acorTop100[i,j]) > p97]
```

```python
igGenesCor = igraph.Graph.TupleList(tlGenesCor, weights = True)
```

```python
igGenesCor
```

```python
igGenesCor.ecount()
```

```python
igGenesCor.vcount()
```

```python
igGenesCor.es['weight']
```

```python
igGenesCor.vs['name']
```

```python
igGenesCor.vs['color'] = '#73b360cc'

igGenesCor.es['color'] = ['#FCCC06' if e['weight'] < 0 else '#007B7F' for e in igGenesCor.es]
```

```python
igraph.plot(igGenesCor, vertex_frame_color = '#00000000')
```
